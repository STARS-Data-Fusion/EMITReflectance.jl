```julia
using HDF5
```

```julia
filename = joinpath(expanduser("~"), "Downloads", "EMIT_L2A_RFL_001_20240521T232728_2414216_003.nc")
```

```julia
GLT_x = h5open(filename, "r") do file
    permutedims(read(file["location/glt_x"]), (2, 1))
end
```

```julia
GLT_y = h5open(filename, "r") do file
    permutedims(read(file["location/glt_y"]), (2, 1))
end
```

```julia
# Stack GLT_x and GLT_y along the third dimension (axis=-1 equivalent)
GLT_array = cat(GLT_x, GLT_y, dims=3)

# Convert to integer type
GLT_array = convert(Array{Int, 3}, GLT_array)
```
